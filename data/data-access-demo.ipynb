```python
# Import driveanon
import driveanon as da
```

First, find the "blob id" for a file in the data [readme](https://github.com/waterhackweek/whw2019_snowmelt/blob/master/data/README.md).

(Or right-click on a file in Google Drive, select "Get sharable link", and use the id following "...?id=" in the url.)

```python
# 30 m resolution digital elevation model (DEM)
# URL to file: https://drive.google.com/open?id=1GTVhP19VF-srNxF50cZH9gTY7yz0I402
blob_id = '1GTVhP19VF-srNxF50cZH9gTY7yz0I402'
```

```python
# Save the file locally (with no filename specified, the original filename is used)
filename = da.save(blob_id, filename=None, overwrite=True)
```

```python
# Open this file with rasterio
import rasterio as rio
dem_file = rio.open(filename)
dem = dem_file.read(1)
```

```python
# Plot the DEM
%matplotlib inline
from matplotlib import pyplot as plt
f,ax = plt.subplots(figsize=(10,10))
ax.imshow(dem, cmap='terrain')
```
