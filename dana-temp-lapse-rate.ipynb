```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import datetime as dt
```

```python
dana = pd.read_csv('data/dana_meadows.csv')
```

```python
type(dana)
```

```python
dana.columns
```

```python
dana['time'] = pd.to_datetime(dana['time'])
```

```python
type(dana['time'][0])
```

```python
start = dt.datetime(2016,1,1)
end = dt.datetime(2016,5,1)
```

```python
dana_2016 = dana[(dana['time']>start) & (dana['time']<end)]
```

```python
dana_2016
```

```python
dana_2016.plot(x='time',y='temp');
```

```python
l_rate = -6.5 # assuming a lapse rate of -6.5 C per 1000 m in elevation
lapse_rate = l_rate/1000# -C per m
melt_band_lower = np.array( dana_2016['temp'] / lapse_rate ) + 3981 # elevation where it is 0 C
melt_band_upper = np.array( (dana_2016['temp']-2) / lapse_rate ) + 3981 # elevation where it is 2 C
```

```python
plt.figure(figsize=(15,5))
plt.plot(dana_2016['time'],melt_band_upper)
plt.plot(dana_2016['time'],melt_band_lower)
plt.xlabel('Time')
plt.ylabel('Snowmelt Elevation Band (0-2 C)')
```
